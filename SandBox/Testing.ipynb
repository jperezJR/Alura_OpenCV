```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from skimage.color import rgb2gray
```

```python
image = plt.imread('1.jpeg')
image.shape
plt.imshow(image)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
```

```python
gray.shape
```

```python
gray_normalized = gray/255
gray_normalized.shape
```

```python
gray_r = gray_normalized.reshape(gray_normalized.shape[0]*gray_normalized.shape[1])
```

```python
gray_r.shape
```

```python
gray_r.mean()
```

```python
for i in range(gray_r.shape[0]):
    if gray_r[i] > 0.55:
        gray_r[i] = 1
    else:
        gray_r[i] = 0
gray = gray_r.reshape(gray.shape[0],gray.shape[1])
plt.imshow(gray, cmap='gray')
```
